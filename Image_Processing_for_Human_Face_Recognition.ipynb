Normalising the data

```
X = (X - X.min())/(X.max()-X.min())
X = X - X.mean()
```

PCA implementation on the data

```
from sklearn.decomposition import PCA

pca = PCA()
pca.fit(X)
X_trans = pca.transform(X)
```

Getting the array of eigen values and eigenvectors





```
eigenvalues= pca.explained_variance_
eigenvectors = pca.components_
#eigenvalues = pca.singular_values_
```

Plot of first 'k' principle components

```
k=np.arange(0,X.shape[0],1)
plt.plot(k,eigenvalues)
plt.ylabel('Eigenvalues')
plt.title('Plot of Eigen values')
plt.axis([-2,X.shape[0],-5,1650])
plt.show()
```

The below cell when executed will give the number of components required to capture 50% energy

```
total = sum(eigenvalues)
add =0
ev = np.zeros((10,len(l)))
for i in range(0,len(eigenvalues),1):
  add = add + eigenvalues[i]
  percent = add/total
  if percent <= 0.5:
    value=i+2
for i in range(0,10,1): 
  ev[i] = ((1/(i+0.5))*l)+eigenvectors[i]
  
  
print('The number of eigen values needed to capture for 50% variance is ' + str(value))
```

Reconstruction of images using 1,10,20,30,40,50 components for subject 4 and 3

```
subject = 4
mat1_mean = mat[subject]+mat[subject+15]+mat[subject+30]+mat[subject+45]+mat[subject+60]+mat[subject+75]+mat[subject+90]+mat[subject+105]+mat[subject+120]+mat[subject+135]+mat[subject+150]
#plt.subplots( sharex=True)
for ev in [1,10,20,30,40,50]:
  c = np.zeros(ev)
  ext=np.zeros(77760)
  for i in range(0,ev,1):
    c[i]=np.dot(eigenvectors[i],mat[subject+15])
    ext = ext + np.transpose(c[i]*eigenvectors[i])
  fin= ((1/11)*mat1_mean) + ext
  fin= np.reshape(fin,(243,320))
  #print('The below image is constructed with ' + str(ev) + ' component(s)')  
  plt.figure(figsize=(3,3))
  plt.imshow(fin.real, cmap='gray', vmin=0, vmax=255)
  plt.show() 
```

5-fold cross validation for different PCA components with different methods

```
PLA_comp = [1,5,10,20,30]
q=0
k=0
i=0
n=5
error_val = np.zeros(len(PLA_comp)*n)
error=np.zeros(len(PLA_comp))
y= np.arange(1,16,1)
y=np.tile(y,11)
for p in PLA_comp:
  pca= PCA(n_components=p)
  pca.fit(X)
  Xh = pca.transform(X)
  X_tr, X_t, y_tr, y_t = train_test_split(Xh, y, test_size=3/11, stratify=y)
  
  skf=StratifiedKFold(n_splits=n, shuffle=True)
  for train_index, test_index in skf.split(X_tr, y_tr):
    X_train, X_test = X_tr[train_index], X_tr[test_index]
    y_train, y_test = y_tr[train_index], y_tr[test_index]
    X_train= X_train.astype('int')
    
    y_train =y_train.astype('int')
    y_test =y_test.astype('int')

    svm = SVC(C=0.1, kernel="rbf", gamma= 'scale')
              
    svm.fit(X_train, y_train)

    pred   =   svm.predict(X_test) 
    
    for j in range(0,len(y_test)):
      if pred[j] != y_test[j]:
        k=k+1
    error_val[q] = k/len(y)
    error[i] = error[i] + k/(n*len(y))
    q=q+1
    k=0
  i=i+1
for i in range(0,len(PLA_comp),1):
  print('The accuracy in SVM for number of PCA component=' + str(PLA_comp[i]) + ' is ' + str(1-error[i]))
```
